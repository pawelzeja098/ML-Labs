```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
irisRaw = load_iris()

iris = pd.DataFrame(data= np.c_[irisRaw['data'], irisRaw['target']],
                    columns= irisRaw['feature_names'] + ['target'])
```

4. Analiza działania groupby()

Grupowanie po sepal length (cm) i liczenie średniej

8. Znaleznienie min i max wartości długości płatka

```python
min_index = iris['petal length (cm)'].idxmin()
max_index = iris['petal length (cm)'].idxmax()

min_value = iris.loc[min_index, 'petal length (cm)']
max_value = iris.loc[max_index, 'petal length (cm)']

print(f"Min wartość dla komulny 3:{min_value} pod indeksem {min_index}")
print(f"Max wartość dla komulny 3:{max_value} pod indeksem {max_index}")
```

9.Odchylenie standardowe

```python
i = 1
for col in iris.columns:
    print(f"Kolumna {i}: {iris[col].std()}")
    i += 1
```

10.


```python

mean_sepal_length = iris['sepal length (cm)'].mean()


filtered_iris = iris[iris['sepal length (cm)'] > mean_sepal_length]

print(filtered_iris)
```

11. Wykres słupkowy przedstawiający średnią długość płatków dla każdej klasy kwiatu.


```python
i = 1
y = []
x = []
for col in iris.columns:
    if col == 'target':
        continue
    y.append(iris[col].mean())
    x.append(col)
    i += 1

plt.bar(x,y)
```
