```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
X, y= make_blobs(n_samples=210, centers=3, n_features=2, cluster_std=0.5, shuffle=True,random_state=0)
```

```python
plt.scatter(X[:,0], X[:,1], c='red', marker='x')
```

```python
km = KMeans(n_clusters = 3,init='random')
km.fit_predict(X)
a = 0
```

```python
plt.figure(figsize=(6,5))
plt.scatter(X[:,0], X[:,1], s=70, c=km.labels_, cmap=plt.cm.prism) 
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], marker='*', s=200,
            color='blue', label='Centers')
plt.legend(loc='best')
plt.xlabel('X0')
plt.ylabel('X1')
```

```python
print(f"Warto znieksztace: {km.inertia_}")
```

Zad 3. k-means++

```python
X, y= make_blobs(n_samples=2500, centers=5, n_features=2, cluster_std=0.7, shuffle=True, )
km_random = KMeans(n_clusters = 5,init='random',n_init = 50)
km_random.fit_predict(X,y)

km_kmeans = KMeans(n_clusters = 5,init='k-means++',n_init = 50)
km_kmeans.fit_predict(X,y)
print(f"Warto znieksztace: {km_random.inertia_}")

plt.figure(figsize=(6,5))
plt.scatter(X[:,0], X[:,1], s=70, c=km_random.labels_, cmap=plt.cm.prism) 
plt.scatter(km_random.cluster_centers_[:, 0], km_random.cluster_centers_[:, 1], marker='*', s=200,
            color='blue', label='Centers')
plt.legend(loc='best')
plt.xlabel('X0')
plt.ylabel('X1')
plt.title('KMeans init = random')
```

```python
plt.figure(figsize=(6,5))
plt.scatter(X[:,0], X[:,1], s=70, c=km_kmeans.labels_, cmap=plt.cm.prism) 
plt.scatter(km_kmeans.cluster_centers_[:, 0], km_kmeans.cluster_centers_[:, 1], marker='*', s=200,
            color='blue', label='Centers')
plt.legend(loc='best')
plt.xlabel('X0')
plt.ylabel('X1')
plt.title('KMeans init = k-means++')
print(f"Warto znieksztace: {km_kmeans.inertia_}")
```

Zadanie 4 - Metoda okciowa

```python
k_range = np.arange(1, 11)

# Tablica na wartoci znieksztace
inertia_values = []

# Oblicz inertia dla rnych k
for k in k_range:
    km = KMeans(n_clusters=k, init='k-means++', n_init=10, random_state=0)
    km.fit_predict(X)
    inertia_values.append(km.inertia_)

# Rysuj wykres
plt.plot(k_range, inertia_values, marker='o')
plt.xlabel('Liczba klastrw (k)')
plt.ylabel('Znieksztacenie (inertia)')
plt.title('Metoda okcia - wybr optymalnej liczby klastrw')
plt.show()
```

```python
from sklearn.datasets import make_moons
import seaborn as sns
 
x, y = make_moons(1000, noise=.05, random_state=0)
X_moon = pd.DataFrame(x, columns = ['f1','f2'])
#metoda klasteryzacji
# km -model KMeans
#y_km - wynik predykcji

km = KMeans(n_clusters = 2,init='k-means++')

y_km = km.fit_predict(x)

#wykres
X_moon['k_means'] = y_km
sns.lmplot(data=X_moon, x='f1', y='f2', fit_reg=False, hue = 'k_means', palette = ['#eb6c6a', '#6aeb6c']).set(title='Algorytm k-srednich')
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], c='black', s=100, alpha=0.5)
plt.show()
```
